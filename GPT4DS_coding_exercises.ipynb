# Exercise 2

The following cell contains uncommented MATLAB code.

Have ChatGPT translate this code into Python and explain what the code does and what data science concepts it illustrates.

```python
n = 40;
x = logspace(log10(1),log10(10),n)+randn(1,n);
y = log(x) + randn(size(x))/5;

p = polyfit(x,y,2);
q = polyval(p,x);
r2 = corr(y(:),q(:))^2;

figure(1), clf
plot(x,y,'ko','markersize',10,'markerfacecolor','k')
hold on
plot(x,q,'rs','markersize',10,'markerfacecolor','r')
xlabel('Predictor')
ylabel('Outcome')
legend({'Data','Predicted'})
title(sprintf('R^2 fit = %g',r2))
```
