```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
N=1000
t=np.linspace(0,2*np.pi,N)
d=np.array((np.cos(t),np.sin(t)))+np.random.randn(2,N)/15
dd=.3*np.array((np.cos(t),np.sin(t)))+np.random.randn(2,N)/15
d=np.hstack((d,dd)).T
q=DBSCAN(eps=.2,min_samples=6).fit(d)
k=q.labels_;j =max(k)+1;l= 'rkbgm'
for i in range(j): plt.plot(d[k==i,0],d[k==i,1],'o',color=l[i%len(l)],markerfacecolor='w')
plt.plot(d[k==-1,0],d[k==-1,1],'k+'),plt.axis('square')
plt.title('Result of dbscan clustering (k=' + str(j) + ')');
```
